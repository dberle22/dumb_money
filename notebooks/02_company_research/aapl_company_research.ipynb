# AAPL Research Memo

This notebook is a thin memo-style review layer over the shared `dumb_money` research and report helpers. It keeps the logic in `src/dumb_money/...` and only assembles the currently supportable visuals and narrative for the `AAPL` workflow.

## Workflow

1. Build the shared `AAPL` research packet.
2. Review the summary strip and memo interpretation.
3. Move through market performance, trend and risk, and current-state business panels.
4. Close with score decomposition and a short research takeaway.

Sections that still need peer modeling or historical fundamentals are called out directly instead of being approximated in-notebook.
